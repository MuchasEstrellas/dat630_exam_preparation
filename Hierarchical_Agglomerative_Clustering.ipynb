# Hierarchical_Agglomerative_Clustering

```python
import pandas as pd
import numpy as np
import pprint
```

```python
def print_table(col,table):
    data = []
    for i, val in enumerate(col):
        data.append([val,table[i]])
    return pd.DataFrame.from_items(data,
                     orient='index', columns=col)
def fill_table(table):
    for i in range(len(table)):
        for j in range(len(table[i])):
            table[j][i] = table[i][j]
    return table
def merge(col,table):

    min_set={"min_val":9999999,"min_col":0,"min_row":0,}
    for i in range(0,len(table)):
        for j in range(0,len(table[i])):
            if i==j : continue
            if min_set["min_val"] > table[i][j]:
                min_set["min_val"] = table[i][j]
                min_set["min_col"]=j
                min_set["min_row"]=i
    tohere = min(min_set["min_col"],min_set["min_row"])
    fromhere = max(min_set["min_col"],min_set["min_row"])
    col[tohere]=str(col[tohere])+"_"+str(col[fromhere])
    col.remove(col[fromhere])
    merge_data = [min(table[i][tohere],table[i][fromhere]) for i in range(len(table[tohere]))]
    table[tohere] = merge_data
    
    for i,val_list in enumerate(table):
        table[i][tohere] = merge_data[i]
        
    for i,val_list in enumerate(table):
        del table[i][fromhere]
    del table[fromhere]        
            
    return col, table
```
